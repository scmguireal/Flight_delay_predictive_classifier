## Checking Passengers DataFrame

```python
df_pass_origin = pd.read_csv('../../../data/passengers_top_origin18_19.csv')
df_pass_origin.head()
```

```python
# Set in order for easy viewing...Kept useful columns
df_pass = df_pass_origin[['year', 'month','origin_airport_id','origin_city_name','dest_city_name','dest_airport_id','departures_performed','seats','passengers','distance','airline_id','unique_carrier','unique_carrier_name', 'carrier_group_new','aircraft_type',
       'aircraft_config',  'distance_group', 'class']]
df_pass.head()
```

```python
# Tables holding values to to link ids with names
carrier_link = df_pass[['airline_id', 'unique_carrier_name', 'unique_carrier']]
airport_link = df_pass[['origin_airport_id', 'origin_city_name','dest_airport_id','dest_city_name']]
```

```python
model_df = df_pass.copy()
```

```python
model_df = model_df.drop(['distance','origin_city_name','dest_city_name','unique_carrier_name'],axis=1)
```

```python
model_df = pd.get_dummies(model_df,prefix=['class'], columns = ['class'], drop_first=True)
```

```python
model_df = model_df.sort_values(['origin_airport_id', 'year','month','airline_id'])
```

```python
model_df = model_df.loc[model_df['unique_carrier'].isin(['WN', 'AA', 'UA', 'DL', 'B6', 'NK', 'G4', 'AS', 'F9', 'HA', 'VX'])]
```

```python
model_df.columns
```

```python
model_df['percent_filled'] = (model_df['passengers']/model_df['seats'])*100
model_df['mkt_unique_carrier'] = model_df['unique_carrier']
```

```python
model_df = model_df[['origin_airport_id', 'dest_airport_id', 'mkt_unique_carrier', 'airline_id', 'month', 'year',
       'departures_performed', 'seats', 'passengers', 'percent_filled',
       'carrier_group_new', 'aircraft_type', 'aircraft_config', 
        'distance_group', 'class_G', 'class_L', 'class_P']]
```

```python
model_df = model_df.sort_values(['origin_airport_id', 'mkt_unique_carrier', 'month','dest_airport_id'])
```

```python
model_df
```

```python
#model_df.loc[(model_df['origin_airport_id'] == 10397) & (model_df['airline_id'] == 19393)]

model_df = model_df.drop(['unique_carrier'], axis=1)
```
